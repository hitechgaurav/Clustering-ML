```python
#Importing the Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as pt
import seaborn as sns
sns.set()
```

```python
data = pd.read_csv("E:\Python All in One\DATA_SETS\Machine Learning A-Z Template Folder\Part 4 - Clustering\Section 25 - Hierarchical Clustering\Mall_Customers.csv")
```

```python
#Targetting our data with topic of intrest
x = data.iloc[:, [3, 4]].values
```

```python
#Finding optimal number of clusters using Dendrogram Method
pt.figure(figsize = (10,5))
import scipy.cluster.hierarchy as sch
z = sch.linkage(x, metric = 'euclidean', method = 'ward')
dendrogram = sch.dendrogram(z)
pt.title("Dendogram")
pt.xlabel("Clusters")
pt.ylabel("Euclidean Distance")
pt.show()
```

```python
#Building our model
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters = 5, affinity = 'euclidean', linkage = 'ward')

cluster = hc.fit_predict(x)
print(cluster)
```
